```python
import numpy as np 
import pandas as pd
import calendar
import matplotlib.pyplot as plt

####Importing CSV Files as dataframes
df = pd.read_csv('chicago_crime_data_v3.csv')
df_prime=pd.read_csv('primary_type_ID.csv')

#####Sorting the crime data and merging both the dataframes using Primary Type ID
df.sort_values(by=['Cime_ID'])
frames = [df, df_prime]
df_crime = pd.merge(df, df_prime, on='Primary Type ID')
df_crime.head()

####Part 3
crimes = df_crime[['Primary Type ID']]
crime_count = pd.DataFrame(crimes.groupby('Primary Type ID').size().sort_values(ascending=False).rename('counts').reset_index())
crime_count
##Thus we notice that the crimes with primary type IDS '32','3','7','2','9','25','4','18','29','17','8' are the high frequency crimes.


##Adding month and year columns to the dataframe
df_crime['Date'] = pd.to_datetime(df_crime['Date']) 
df_crime['Month']=(df_crime['Date'].dt.month).apply(lambda x: calendar.month_abbr[x]) 
df_crime['Year']=(df_crime['Date'].dt.year)
df_crime.head()

##Plotting counts for each high frequency crime each month from Jan 2016-Dec 2017
plt.figure(figsize=(14,8))
ID32 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==32])
series32=ID32.groupby(['Month','Year'])['Primary Type ID'].count()
series32.plot(label='Primary Type ID=32')

ID3 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==3])
series3=ID3.groupby(['Month','Year'])['Primary Type ID'].count()
series3.plot(label='Primary Type ID=3')

ID7 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==7])
series7=ID7.groupby(['Month','Year'])['Primary Type ID'].count()
series7.plot(label='Primary Type ID=7')

ID2 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==2])
series2=ID2.groupby(['Month','Year'])['Primary Type ID'].count()
series2.plot(label='Primary Type ID=2')

ID9 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==9])
series9=ID9.groupby(['Month','Year'])['Primary Type ID'].count()
series9.plot(label='Primary Type ID=9')

ID25 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==25])
series25=ID25.groupby(['Month','Year'])['Primary Type ID'].count()
series25.plot(label='Primary Type ID=25')

ID4 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==4])
series4=ID4.groupby(['Month','Year'])['Primary Type ID'].count()
series4.plot(label='Primary Type ID=4')

ID18 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==18])
series18=ID18.groupby(['Month','Year'])['Primary Type ID'].count()
series18.plot(label='Primary Type ID=18')

ID29 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==29])
series29=ID29.groupby(['Month','Year'])['Primary Type ID'].count()
series29.plot(label='Primary Type ID=29')

ID17 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==17])
series17=ID17.groupby(['Month','Year'])['Primary Type ID'].count()
series17.plot(label='Primary Type ID=17')

ID8 = pd.DataFrame(df_crime[df_crime['Primary Type ID']==32])
series8=ID32.groupby(['Month','Year'])['Primary Type ID'].count()
series8.plot(label='Primary Type ID=8')


months=('Jan-16', 'Feb-16','Mar-16','Apr-16','May-16','Jun-16','Jul-16','Aug-16','Sep-16','Oct-16','Nov-16','Dec-16','Jan-17', 'Feb-17','Mar-17','Apr-17','May-17','Jun-17','Jul-17','Aug-17','Sep-17','Oct-17','Nov-17','Dec-17')
plt.xticks(np.arange(24),months, rotation=90)
plt.xlabel('Jan 2016 - Dec 2017')
plt.ylabel('Number of high Frequency crimes')
plt.title('High frequency crimes in 2016 and 2017 timeseries plot')
plt.legend() 
plt.show()


```
